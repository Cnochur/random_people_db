```python
import requests
import json
import pandas as pd
import random
```

This programme uses the randomuser.me API to generate 100 random people with some details

```python
def get_random_person():

    base_site = 'https://randomuser.me/api'
    response = requests.get(base_site)

    title = response.json()['results'][0]['name']['title']
    first_name = response.json()['results'][0]['name']['first']
    last_name = response.json()['results'][0]['name']['last']
    gender = response.json()['results'][0]['gender']
    street = response.json()['results'][0]['location']['street']['name']
    country = response.json()['results'][0]['location']['country']
    age = random.randint(18,86)

    return{'Title':title,'First Name': first_name, 'Surname' : last_name, 'Sex': gender, 'Street': street, 'Country': country, 'Age': age}
    

#Create a list of dictionaries containg information
def get_list_of_people():

    random_list = []

    while len(random_list) < 100:
        new_person = get_random_person()
        random_list.append(new_person)

    return random_list


```

```python
#creating a list to create DataFrame
list_one = get_list_of_people()

db = pd.DataFrame(list_one)
db
```
